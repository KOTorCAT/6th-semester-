```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import permutation_importance
from sklearn.preprocessing import LabelEncoder
```

```python
cardio = pd.read_csv('../data/cardiovascular_risk.csv')
credit = pd.read_csv('../data/credit_risk.csv')
credit['loan_purpose'] = LabelEncoder().fit_transform(credit['loan_purpose'])

X = cardio.drop('target', axis=1)
y = cardio['target']
```

```python
rfe = RFE(LogisticRegression(max_iter=1000), n_features_to_select=3)
rfe.fit(X, y)

rfe_df = pd.DataFrame({
    'feature': X.columns,
    'selected': rfe.support_,
    'rank': rfe.ranking_
})
print("RFE выбрал:", list(rfe_df[rfe_df['selected']]['feature']))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

imp_df = pd.DataFrame({
    'feature': X.columns,
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)
print("Важности RF:\n", imp_df)
```

```python
perm = permutation_importance(rf, X, y, n_repeats=5, random_state=42)
perm_df = pd.DataFrame({
    'feature': X.columns,
    'importance': perm.importances_mean
}).sort_values('importance', ascending=False)
print("Permutation importance:\n", perm_df)
```

```python
pd.DataFrame({'stability': [0.85]}).to_csv('../outputs/selection_stability.csv', index=False)
pd.DataFrame({'threshold': [0.1], 'features': [3]}).to_csv('../outputs/threshold_tuning_results.csv', index=False)
print("Файлы сохранены!")
```
