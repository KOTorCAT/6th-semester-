```python
# Ячейка 1
"""
## Муниципальная пекарня

Пекарня выпекает хлеб (X) и булочки (Y).
Ресурсы:
- Мука: 2 кг на хлеб, 1 кг на булочку, всего 100 кг
- Время: 1 час на хлеб, 2 часа на булочку, всего 80 часов
- Спрос: булочек не более 40 штук
Прибыль: хлеб - 50 руб, булочка - 30 руб

Задание:
1. Записать математическую модель
2. Построить допустимую область
3. Найти вершины и прибыль в них
4. Решить через linprog
5. Объяснить результат
"""
```

```python
# Ячейка 2: Модель
"""
Математическая модель:
max Z = 50X + 30Y

Ограничения:
2X + Y <= 100  (мука)
X + 2Y <= 80   (время)
Y <= 40        (спрос)
X >= 0, Y >= 0
"""
```

```python
# Ячейка 3: Импорты
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
```

```python
# Ячейка 4: График
x = np.linspace(0, 60, 100)
y1 = 100 - 2*x
y2 = (80 - x) / 2
y3 = np.full_like(x, 40)

plt.figure(figsize=(8, 6))
plt.fill_between(x, 0, np.minimum(np.minimum(y1, y2), y3), 
                 where=(x>=0) & (y1>=0) & (y2>=0), alpha=0.3)
plt.plot(x, y1, 'b-', label='Мука: 2X+Y=100')
plt.plot(x, y2, 'r-', label='Время: X+2Y=80')
plt.axhline(y=40, color='g', label='Спрос: Y=40')

# Вершины
vertices = [(0,0), (0,40), (20,40), (40,20), (50,0)]
for vx, vy in vertices:
    if vy >= 0 and 2*vx+vy <= 100 and vx+2*vy <= 80 and vy <= 40:
        plt.plot(vx, vy, 'ko', markersize=10)
        Z = 50*vx + 30*vy
        print(f"Вершина ({vx}, {vy}): Z = {Z}")

plt.xlim(0, 60)
plt.ylim(0, 50)
plt.xlabel('Хлеб (X)')
plt.ylabel('Булочки (Y)')
plt.title('Допустимая область')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ячейка 5: linprog
c = [-50, -30]
A = [[2, 1], [1, 2], [0, 1]]
b = [100, 80, 40]
result = linprog(c, A_ub=A, b_ub=b, bounds=(0, None), method='highs')

print(f"Оптимальный план: X={result.x[0]:.1f}, Y={result.x[1]:.1f}")
print(f"Максимальная прибыль: {-result.fun:.1f} руб")
```
