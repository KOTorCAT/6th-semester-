```python
# Ячейка 1
"""
## Городской ремонтный участок

Бригады X и Y выполняют ремонт дорог и тротуаров.
Бригада X: 3 км дорог + 1 км тротуаров в день, оплата 15 тыс/день
Бригада Y: 1 км дорог + 2 км тротуаров в день, оплата 12 тыс/день
План: не менее 18 км дорог и 12 км тротуаров
Цель: минимизировать затраты

Задание:
1. Записать математическую модель
2. Построить допустимую область
3. Решить через linprog
"""
```

```python
# Ячейка 2: Модель и решение
"""
min Z = 15X + 12Y
3X + Y >= 18
X + 2Y >= 12
X >= 0, Y >= 0
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

x = np.linspace(0, 15, 100)
y1 = 18 - 3*x
y2 = (12 - x) / 2

plt.figure(figsize=(8, 6))
plt.fill_between(x, np.maximum(y1, y2), 15, where=(x>=0)&(y1>=0)&(y2>=0), alpha=0.3)
plt.plot(x, y1, 'b-', label='Дороги: 3X+Y=18')
plt.plot(x, y2, 'r-', label='Тротуары: X+2Y=12')

c = [15, 12]
A = [[-3, -1], [-1, -2]]
b = [-18, -12]
result = linprog(c, A_ub=A, b_ub=b, bounds=(0, None), method='highs')

plt.plot(result.x[0], result.x[1], 'ko', markersize=10)
print(f"X={result.x[0]:.1f}, Y={result.x[1]:.1f}, Затраты={result.fun:.1f} тыс")

plt.xlim(0, 15)
plt.ylim(0, 15)
plt.legend()
plt.grid(True)
plt.show()
```
