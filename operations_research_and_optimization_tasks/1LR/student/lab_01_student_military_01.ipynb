```python
# Ячейка 1
"""
## Комплектование полевых рационов

Составить рацион из продуктов A и B.
Питательные вещества:
- Белки: A=4, B=2, минимум 20
- Жиры: A=1, B=3, минимум 15
- Углеводы: A=2, B=5, минимум 30
Стоимость: A=10, B=8 руб/ед

Задание:
1. Записать математическую модель
2. Построить допустимую область
3. Найти вершины и стоимость в них
4. Решить через linprog
5. Объяснить результат
"""
```

```python
# Ячейка 3: График
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

a = np.linspace(0, 10, 100)
b1 = (20 - 4*a) / 2
b2 = (15 - a) / 3
b3 = (30 - 2*a) / 5

plt.figure(figsize=(8, 6))
plt.fill_between(a, np.maximum(np.maximum(b1, b2), b3), 10, alpha=0.3)
plt.plot(a, b1, 'b-', label='Белки: 4A+2B=20')
plt.plot(a, b2, 'r-', label='Жиры: A+3B=15')
plt.plot(a, b3, 'g-', label='Углеводы: 2A+5B=30')

vertices = [(3,4), (2.5,5), (15,0)]
for va, vb in vertices:
    Z = 10*va + 8*vb
    print(f"A={va}, B={vb}: Z = {Z}")

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('Продукт A')
plt.ylabel('Продукт B')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ячейка 4: linprog
c = [10, 8]
A = [[-4, -2], [-1, -3], [-2, -5]]
b = [-20, -15, -30]
result = linprog(c, A_ub=A, b_ub=b, bounds=(0, None), method='highs')

print(f"Оптимальный рацион: A={result.x[0]:.1f}, B={result.x[1]:.1f}")
print(f"Минимальная стоимость: {result.fun:.1f} руб")
```
