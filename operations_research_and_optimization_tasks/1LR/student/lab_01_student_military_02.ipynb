```python
# Ячейка 1
"""
## Ремонтно-обслуживающий склад

Склад хранит запчасти типов X и Y.
Тип X: занимает 2 м3, обслуживание 100 руб/мес
Тип Y: занимает 3 м3, обслуживание 80 руб/мес
Объем склада: 120 м3
Потребность: не менее 20 единиц X и 30 единиц Y
Цель: минимизировать затраты на обслуживание

Задание:
1. Записать модель
2. Решить через linprog
3. Проверить использование склада
"""
```

```python
# Ячейка 2: Решение
"""
min Z = 100X + 80Y
2X + 3Y <= 120
X >= 20
Y >= 30
"""

from scipy.optimize import linprog

c = [100, 80]
A = [[2, 3]]
b = [120]
bounds = [(20, None), (30, None)]
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

print(f"X={result.x[0]:.1f}, Y={result.x[1]:.1f}")
print(f"Затраты={result.fun:.1f} руб")
print(f"Заполнение склада: {2*result.x[0] + 3*result.x[1]:.1f} м3 из 120")

"""
Ответ: X=20, Y=26.7.
Затраты: 4133 руб.
Склад заполнен полностью.
"""
```
