```python
# Ячейка 2: Проверка баланса и модель
import numpy as np
from scipy.optimize import linprog

a = [50, 70]  # запасы
b = [40, 30, 50]  # спрос

print(f"Запасы: {sum(a)}, Спрос: {sum(b)}")
print(f"Баланс: {'закрытая' if sum(a) == sum(b) else 'открытая'}")

# Матрица стоимостей
c = np.array([10, 15, 20,  # A->Б1, A->Б2, A->Б3
              12,  8, 18]) # B->Б1, B->Б2, B->Б3
```

```python
# Ячейка 3: Построение A_eq
m, n = 2, 3
A_eq = []

# Ограничения поставщиков
for i in range(m):
    row = np.zeros(m * n)
    row[i*n:(i+1)*n] = 1
    A_eq.append(row)

# Ограничения потребителей
for j in range(n):
    row = np.zeros(m * n)
    row[j::n] = 1
    A_eq.append(row)

A_eq = np.array(A_eq)
b_eq = a + b

print("Матрица ограничений:")
print(A_eq)
print("Правые части:", b_eq)
```
