```python
# Ячейка 1: Постановка
"""
## Школьное питание

Фабрика A (300 порций), Фабрика B (200 порций).
Школа 1 (150), Школа 2 (250), Школа 3 (100).

Стоимость доставки порции:
   Ш1  Ш2  Ш3
A:  5   8   7
B:  6   4   9

Задание:
1. Проверить баланс
2. Решить
3. Какая школа получит питание от какой фабрики
"""
```

```python
# Ячейка 2: Решение
import numpy as np
from scipy.optimize import linprog

a = [300, 200]
b = [150, 250, 100]

print(f"Баланс: {sum(a)} = {sum(b)}")

c = np.array([5,8,7,  6,4,9])

A_eq = []
for i in range(2):
    row = np.zeros(6)
    row[i*3:(i+1)*3] = 1
    A_eq.append(row)
for j in range(3):
    row = np.zeros(6)
    row[j::3] = 1
    A_eq.append(row)

result = linprog(c, A_eq=A_eq, b_eq=a+b, bounds=(0, None), method='highs')
X = result.x.reshape(2, 3)

print(f"A: Ш1={X[0,0]:.0f} Ш2={X[0,1]:.0f} Ш3={X[0,2]:.0f}")
print(f"B: Ш1={X[1,0]:.0f} Ш2={X[1,1]:.0f} Ш3={X[1,2]:.0f}")
print(f"Стоимость: {result.fun:.0f}")
```
