```python
# Ячейка 1: Постановка
"""
## Топливо для частей

База 1 (80 тонн), База 2 (60 тонн), База 3 (40 тонн).
Часть A (50), Часть B (70), Часть C (30), Часть D (30).

Стоимость перевозки тонны (тыс. руб):
    A   B   C   D
1:  5   8  12   6
2:  7   4   9  10
3: 11   6   5   8

Задание:
1. Проверить баланс (открытая задача)
2. Добавить фиктивного потребителя
3. Решить через linprog
4. Объяснить план
"""
```

```python
# Ячейка 2: Модель с фиктивным узлом
import numpy as np
from scipy.optimize import linprog

a = [80, 60, 40]  # запасы
b = [50, 70, 30, 30]  # спрос

print(f"Запасы: {sum(a)}, Спрос: {sum(b)}")
diff = sum(a) - sum(b)
print(f"Разница: {diff}")

if diff > 0:
    b.append(diff)
    c = np.array([5,8,12,6,0,  7,4,9,10,0,  11,6,5,8,0])
elif diff < 0:
    a.append(-diff)
    c = np.array([5,8,12,6,  7,4,9,10,  11,6,5,8,  0,0,0,0])

m, n = 3, 5 if diff > 0 else 4

A_eq = []
for i in range(m):
    row = np.zeros(m * n)
    row[i*n:(i+1)*n] = 1
    A_eq.append(row)
for j in range(n):
    row = np.zeros(m * n)
    row[j::n] = 1
    A_eq.append(row)

b_eq = a + b

result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=(0, None), method='highs')
X = result.x.reshape(m, n)

print(f"Минимальная стоимость: {result.fun:.0f} тыс. руб")
print("Избыток топлива остался на базах (фиктивный потребитель)")
```
