```python
# Ячейка 1: Постановка
"""
## Запчасти на ремонтные базы

Склад 1 (100 ед), Склад 2 (150 ед), Склад 3 (50 ед).
Рембаза X (80), Рембаза Y (120), Рембаза Z (70).

Стоимость доставки единицы:
    X   Y   Z
1:  3   7   5
2:  6   4   8
3:  9   3   6

Задание:
1. Проверить баланс (дефицит)
2. Добавить фиктивного поставщика
3. Какая рембаза недополучит запчасти
"""
```

```python
# Ячейка 2: Решение с дефицитом
import numpy as np
from scipy.optimize import linprog

a = [100, 150, 50]
b = [80, 120, 70]

print(f"Запасы: {sum(a)}, Спрос: {sum(b)}")
diff = sum(b) - sum(a)
print(f"Дефицит: {diff}")

a.append(diff)  # фиктивный поставщик
c = np.array([3,7,5,  6,4,8,  9,3,6,  0,0,0])

A_eq = []
for i in range(4):
    row = np.zeros(12)
    row[i*3:(i+1)*3] = 1
    A_eq.append(row)
for j in range(3):
    row = np.zeros(12)
    row[j::3] = 1
    A_eq.append(row)

result = linprog(c, A_eq=A_eq, b_eq=a+b, bounds=(0, None), method='highs')
X = result.x.reshape(4, 3)

print(f"Стоимость: {result.fun:.0f}")
print(f"Фиктивный поставщик (недопоставки): {X[3,0]:.0f} {X[3,1]:.0f} {X[3,2]:.0f}")
```
