```python
# Ячейка 1: Постановка
"""
## Социальная защита в зимний период

Primal: распределение средств на отопление (X) и питание (Y).
Полезность: 40 ед на отопление, 60 ед на питание.

Ограничения:
- Фонд: 5X + 8Y <= 400
- Волонтеры: 2X + Y <= 100
- Минимум питания: Y >= 20
X, Y >= 0

Задание:
1. Решить primal
2. Найти теневые цены
3. Что выгоднее: +1 волонтер или +10 фонда
4. Проверить прогноз
"""
```

```python
# Ячейка 2: Решение
import numpy as np
from scipy.optimize import linprog

c = [-40, -60]
A = [[5, 8], [2, 1], [0, -1]]
b = [400, 100, -20]

result = linprog(c, A_ub=A, b_ub=b, bounds=(0, None), method='highs')
print(f"X={result.x[0]:.2f}, Y={result.x[1]:.2f}, Z={-result.fun:.2f}")

# Теневые цены из решения (для scipy >= 1.9)
if hasattr(result, 'ineqlin') and hasattr(result.ineqlin, 'marginals'):
    print("Shadow prices:", result.ineqlin.marginals)
else:
    # Решаем dual
    c_dual = [400, 100, -20]
    A_dual = [[-5, -2, 0], [-8, -1, 1]]
    b_dual = [-40, -60]
    dual = linprog(c_dual, A_ub=A_dual, b_ub=b_dual, bounds=(0, None), method='highs')
    print(f"Фонд: {dual.x[0]:.3f}")
    print(f"Волонтеры: {dual.x[1]:.3f}")
    print(f"Мин питание: {dual.x[2]:.3f}")

# Сравнение
print("\n+1 волонтер (цена y2) vs +10 фонда (10 * y1)")
print(f"Волонтер: {dual.x[1]:.3f}")
print(f"10 фонда: {10*dual.x[0]:.3f}")
```
