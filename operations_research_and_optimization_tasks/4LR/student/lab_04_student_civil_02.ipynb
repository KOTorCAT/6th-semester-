```python
# Ячейка 1: Постановка
"""
## Преподаватели по учебным группам

3 преподавателя, 3 группы.
Рейтинг совместимости (баллы, максимизация):
      Г1  Г2  Г3
Преп1: 85  70  90
Преп2: 75  95  80
Преп3: 90  85  75

Задание:
1. Перевести в задачу минимизации
2. Решить через linear_sum_assignment
3. Проверить через linprog
"""
```

```python
# Ячейка 2: Решение (максимизация -> минимизация)
import numpy as np
from scipy.optimize import linear_sum_assignment, linprog

rating = np.array([
    [85, 70, 90],
    [75, 95, 80],
    [90, 85, 75]
])

# Max -> Min: вычитаем из максимума
cost = rating.max() - rating

row_ind, col_ind = linear_sum_assignment(cost)

print("Назначения (преподаватель -> группа):")
total = 0
for i, j in zip(row_ind, col_ind):
    print(f"Преподаватель {i+1} -> Группа {j+1}: рейтинг {rating[i,j]}")
    total += rating[i, j]
print(f"Суммарный рейтинг: {total}")

X = np.zeros_like(cost)
X[row_ind, col_ind] = 1
assert X.sum(axis=1).all() == 1
assert X.sum(axis=0).all() == 1

# LP с максимизацией
c = -rating.flatten()  # min -sum = max sum
n = 3
A_eq = []
for i in range(n):
    row = np.zeros(n*n)
    row[i*n:(i+1)*n] = 1
    A_eq.append(row)
for j in range(n):
    row = np.zeros(n*n)
    row[j::n] = 1
    A_eq.append(row)

result = linprog(c, A_eq=A_eq, b_eq=[1]*(2*n), bounds=(0, 1), method='highs')
print(f"LP рейтинг: {-result.fun:.0f}")
```
