```python
# Ячейка 1: Постановка
"""
## Ремонтные бригады по единицам техники

5 бригад, 5 единиц техники.
Время ремонта (часов):
      Т1  Т2  Т3  Т4  Т5
Бр1:   8  12  10  15   9
Бр2:  14   7  11  10  13
Бр3:  10  15   6  12   8
Бр4:  12   9  14   5  11
Бр5:   9  11   8  13   7

Задание:
1. Найти оптимальное назначение
2. Проверить ограничения
3. Сравнить с LP
"""
```

```python
# Ячейка 2: Решение
import numpy as np
from scipy.optimize import linear_sum_assignment, linprog

cost = np.array([
    [8, 12, 10, 15, 9],
    [14, 7, 11, 10, 13],
    [10, 15, 6, 12, 8],
    [12, 9, 14, 5, 11],
    [9, 11, 8, 13, 7]
])

row_ind, col_ind = linear_sum_assignment(cost)

print("Назначения (бригада -> техника):")
total = 0
for i, j in zip(row_ind, col_ind):
    print(f"Бригада {i+1} -> Техника {j+1}: {cost[i,j]} ч")
    total += cost[i, j]
print(f"Общее время: {total} ч")

# Проверки
X = np.zeros_like(cost)
X[row_ind, col_ind] = 1
assert X.sum(axis=1).all() == 1
assert X.sum(axis=0).all() == 1

# LP проверка
n = 5
c = cost.flatten()
A_eq = []
for i in range(n):
    row = np.zeros(n*n)
    row[i*n:(i+1)*n] = 1
    A_eq.append(row)
for j in range(n):
    row = np.zeros(n*n)
    row[j::n] = 1
    A_eq.append(row)

result = linprog(c, A_eq=A_eq, b_eq=[1]*(2*n), bounds=(0, 1), method='highs')
print(f"LP стоимость: {result.fun:.0f} ч")
print(f"Решения совпадают: {np.allclose(X.flatten(), result.x.round())}")
```
