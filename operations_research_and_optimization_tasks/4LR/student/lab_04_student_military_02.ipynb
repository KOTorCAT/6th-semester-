```python
# Ячейка 1: Постановка
"""
## Экипажи по задачам вылета

4 экипажа, 4 задачи.
Риск выполнения (баллы, минимизация):
      З1  З2  З3  З4
Эк1:  30  45  25  50
Эк2:  40  20  35  30
Эк3:  25  40  45  35
Эк4:  35  30  20  25

Задание:
1. Найти назначение минимального суммарного риска
2. Проверить ограничения
3. LP-проверка
"""
```

```python
# Ячейка 2: Решение
import numpy as np
from scipy.optimize import linear_sum_assignment, linprog

risk = np.array([
    [30, 45, 25, 50],
    [40, 20, 35, 30],
    [25, 40, 45, 35],
    [35, 30, 20, 25]
])

row_ind, col_ind = linear_sum_assignment(risk)

print("Назначения (экипаж -> задача):")
total = 0
for i, j in zip(row_ind, col_ind):
    print(f"Экипаж {i+1} -> Задача {j+1}: риск {risk[i,j]}")
    total += risk[i, j]
print(f"Суммарный риск: {total}")

X = np.zeros_like(risk)
X[row_ind, col_ind] = 1
assert X.sum(axis=1).all() == 1
assert X.sum(axis=0).all() == 1

# LP
n = 4
c = risk.flatten()
A_eq = []
for i in range(n):
    row = np.zeros(n*n)
    row[i*n:(i+1)*n] = 1
    A_eq.append(row)
for j in range(n):
    row = np.zeros(n*n)
    row[j::n] = 1
    A_eq.append(row)

result = linprog(c, A_eq=A_eq, b_eq=[1]*(2*n), bounds=(0, 1), method='highs')
print(f"LP риск: {result.fun:.0f}")
print(f"Совпадают: {np.allclose(X.flatten(), result.x.round())}")
```
